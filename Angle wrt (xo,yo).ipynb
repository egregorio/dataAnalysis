```python
# Set initial value for points 1-3
p1_xo = p1_x[0]
p1_yo = p1_y[0]
p2_xo = p2_x[0]
p2_yo = p2_y[0]
p3_xo = p3_x[0]
p3_yo = p3_y[0]
```

```python
y_t = [p1_y, p2_y, p3_y]
x_t = [p1_x, p2_x, p3_x]

yo = [p1_yo, p2_yo, p3_yo]
xo = [p1_xo, p2_xo, p3_xo]

theta_1 = np.zeros(99)
theta_2 = np.zeros(99)
theta_3 = np.zeros(99)

theta = [theta_1, theta_2, theta_3]

print(np.shape(theta))

for j in range(0,len(y_t)):
    for i in range(1,len(y_t[1][1:])):
        delta_y = yo[j] - y_t[j][i]
        delta_x = xo[j] - x_t[j][i]
        theta_i = np.arctan(delta_y / delta_x)
        theta[j][i] = theta_i
```

```python
plt.figure()
plt.plot(theta[0],label='p1')
plt.plot(theta[1],label='p2')
plt.plot(theta[2],label='p3')
plt.legend()
plt.show()
```
